```python
def create_time_buckets_offer_received(transcript_df, save=None):
    
    '''
    Creates features based on past customer history
    '''

    bar = progressbar.ProgressBar()
  
    transcript_df['hist_reward_completed'] = 0 
    transcript_df['hist_reward_possible'] = 0 
    transcript_df['hist_difficulty_completed'] = 0 
    transcript_df['hist_difficulty_possible'] = 0 
    transcript_df['hist_previous_completed'] = 0 
    transcript_df['hist_previous_offers'] = 0 
    transcript_df['hist_viewed_and_completed'] = 0 
    transcript_df['hist_complete_not_viewed'] = 0 
    transcript_df['hist_reward_possible'] = 0 
    transcript_df['hist_difficulty_possible'] = 0 
    transcript_df['hist_previous_offers'] = 0 
    transcript_df['hist_failed_complete'] = 0
    transcript_df['hist_viewed'] = 0
    transcript_df['hist_received_spend'] = 0
    transcript_df['hist_viewed_spend'] = 0

    for i in bar(transcript_df.index):
 

        for j in transcript_df.index[0:i][::-1]:

                          
            if transcript_df.loc[j, 'person'] != transcript_df.loc[i, 'person']:
                break            
            
            if transcript_df.loc[j, 'completed'] == 1: 

         
            
                transcript_df.loc[i, 'hist_reward_completed'] += transcript_df.loc[j, 'reward']
                transcript_df.loc[i, 'hist_reward_possible'] += transcript_df.loc[j, 'reward']
                
                transcript_df.loc[i, 'hist_difficulty_completed'] += transcript_df.loc[j, 'difficulty']
                transcript_df.loc[i, 'hist_difficulty_possible'] += transcript_df.loc[j, 'difficulty']
                
                transcript_df.loc[i, 'hist_previous_completed'] += 1
                transcript_df.loc[i, 'hist_previous_offers'] += 1
                
                if transcript_df.loc[j, 'viewed'] == 1: 
                    transcript_df.loc[i, 'hist_viewed_and_completed'] += 1
                
                else:
                    transcript_df.loc[i, 'hist_complete_not_viewed'] += 1
                    
                                 
            else:
                transcript_df.loc[i, 'hist_reward_possible'] += transcript_df.loc[j, 'reward']
                transcript_df.loc[i, 'hist_difficulty_possible'] += transcript_df.loc[j, 'difficulty']
                transcript_df.loc[i, 'hist_previous_offers'] += 1
                transcript_df.loc[i, 'hist_failed_complete'] += 1 
                
            if transcript_df.loc[j, 'viewed'] == 1:
                transcript_df.loc[i, 'hist_viewed'] += 1
                
            transcript_df.loc[i, 'hist_received_spend'] += transcript_df.loc[j, 'received_spend']
            transcript_df.loc[i, 'hist_viewed_spend'] += transcript_df.loc[j, 'viewed_spend']
            
    
    if save:
        try:
            dirName='../../data/interim'
            os.mkdir(dirName)
            print("Directory " , dirName ,  " Created ") 
        except FileExistsError:
            pass

        df.to_pickle(dirName + '/testing_hist2.pickle')
        print('saved as {}'.format(dirName + '/testing_hist2.pickle'))        
    
    return transcript_df
```

```python
def label_encode_categories(df):
    '''
    label encodes gender and id, removing previous created one hot encoding    
    
    '''
    le = LabelEncoder()
    df.id = le.fit_transform(df.id)
    le.fit_transform(df.id)
    df['gender'] = df[['F', 'M', 'O']].idxmax(1)
    df.gender = le.fit_transform(df.gender)
    df = df.drop(['F', 'M', 'O'], axis=1)
    return df
```

```python
def last_view(offer_transaction, save=True):
    
    '''
    Creates time bucket fields for total transactions and number of transactions going back in time against the offer reveived event
    simplyifying to correct last transaction and last_amount

    '''
    
    offer_transaction['last_view_date'] = np.nan

    

    bar = progressbar.ProgressBar()

    for i in bar(offer_transaction.index):

        for j in offer_transaction.index[0:i][::-1]:
                          
            if offer_transaction.loc[j, 'person'] != offer_transaction.loc[i, 'person']:
                break  

            if offer_transaction.loc[j, 'viewed_in_valid'] == 1:

                offer_transaction.loc[i, 'last_view_date'] = offer_transaction.loc[j, 'date'] - offer_transaction.loc[i, 'date']

                break
    
    if save:
        try:
            dirName='../../data/interim'
            os.mkdir(dirName)
            print("Directory " , dirName ,  " Created ") 
        except FileExistsError:
            pass

        offer_transaction.to_pickle(dirName + '/final_last_view.pickle')
        print('saved as {}'.format(dirName + '/final_last_view.pickle'))        

    return df
```
